```python
import pandas as pd

train_data = pd.read_csv('C:/Users/Jorge Payà/Desktop/4Geeks/DSML Bootcamp/Boosting-Algorithms/data/processed/clean_train.csv')
test_data = pd.read_csv('C:/Users/Jorge Payà/Desktop/4Geeks/DSML Bootcamp/Boosting-Algorithms/data/processed/clean_test.csv')

train_data.head()
```

```python
# Split the data into training and testing sets
X_train = train_data.drop(['Outcome'], axis=1) # training data without the target variable
y_train = train_data['Outcome'] # training data target variable
X_test = test_data.drop(['Outcome'], axis=1) # testing data without the target variable
y_test = test_data['Outcome'] # testing data target variable
```

<h3 id="step2" style="font-family: Comic Sans MS; color: #68FF33">2. Boosting Algorithim: XGBClassifier</h3>

```python
from xgboost import XGBClassifier

# Create a XGBoost classifier and add parameters
model = XGBClassifier(
    n_estimators=50,
    max_depth=3,
    learning_rate=0.1,  
    subsample=0.5,
    random_state=42,
    objective='binary:logistic'
)

# Train the model
model.fit(X_train, y_train)
```

```python
# Predict the target variable
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

<ol style="color: pink">
    <li>n_estimators=100, max_depth=3, learning_rate=0.1, subsample=0.5, random_state=0 >>>> The result was <b>75.97%</b></li>
